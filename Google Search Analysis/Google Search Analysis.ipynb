```python
# Using pytrends library to see the trends on google searches from Google Trends

import pandas as pd
%pip install pytrends
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

```python
# Looking into the searches of "Cyber Security"
# Interest by Region: returns data for where the keyword
# is most searched as shown on Google Trends' Interest by Region section.

trends.build_payload(kw_list=["Cyber Security"])
data = trends.interest_by_region()
data = data.sort_values(by="Cyber Security", ascending=False)
data = data.head(10)
print(data)


```

```python
#Visualizing the interest by region with a bar chart

data.reset_index().plot(x="geoName", y="Cyber Security", 
                        figsize=(10,10), kind="bar")
plt.style.use('seaborn-dark-palette')
plt.title('Interest by Region of Google Searches for Cyber Security', fontweight='bold')

plt.show()
```

```python
#checking what people are searching in switzerland

trends.trending_searches(pn='switzerland').head(10)
```

```python
#checking what people are searching in switzerland - real time

trends.realtime_trending_searches(pn='CH').head(10)
```

```python
# Finding out how the total searches for the phrase we
# are looking into has changed over time

# Connecting to Google and building the payload
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Cyber Security'])
data = data.interest_over_time()

figure, axes = plt.subplots(figsize=(10, 10))
data['Cyber Security'].plot()
plt.style.use('seaborn-dark-palette')
plt.title('Total Google Searches for Cyber Security', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()

```
